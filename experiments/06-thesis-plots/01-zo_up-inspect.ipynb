```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
# Replace 'your_file.csv' with the actual filename or path
df = pd.read_csv(DATA_DIR_PATH / "zo_up.csv")

# Count the occurrences of each SDG class
sdg_counts = df['sdg'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=sdg_counts.index, y=sdg_counts.values)

# Customize the plot
plt.title('Distribution of SDG Classes', fontsize=16)
plt.xlabel('SDG Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add value labels on top of each bar
for i, v in enumerate(sdg_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
